```python
from scipy.stats import skew
# Create empty lists to store the column names
skewed_cols = []
normal_cols = []

# Loop through each column in the dataframe
for col in df.columns:
    # Check if the column is numeric
    if df[col].dtype in ['int64', 'float64']:
        # Calculate the skewness of the column
        skewness = skew(df[col])
        # If the skewness is greater than 1 or less than -1, it's considered skewed
        if skewness > 1 or skewness < -1:
            skewed_cols.append(col)
        else:
            normal_cols.append(col)

print('Skewed Columns:', skewed_cols)
print('Normal Columns:', normal_cols)
```

```python
from scipy.stats import zscore


# detect and remove outliers in skewed columns using IQR
for col in skewed_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]

# detect and remove outliers in normal columns using Z-score
for col in normal_cols:
    z_scores = zscore(df[col])
    df = df[np.abs(z_scores) <= 2]

print('Dataframe without outliers')
```

```python
# Calculate the IQR for skewed columns
Q1 = df[skewed_cols].quantile(0.25)
Q3 = df[skewed_cols].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers from skewed columns
df = df[~((df[skewed_cols] < (Q1 - 1.5 * IQR)) | (df[skewed_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Calculate the z-score for normal columns
z = np.abs((df[normal_cols] - df[normal_cols].mean()) / df[normal_cols].std())

# Remove outliers from normal columns
df = df[(z < 3).all(axis=1)]
```

```python

from scipy.stats import zscore

# assuming df is the pandas dataframe containing the data

# detect outliers in skewed columns using IQR
outliers_skewed = pd.DataFrame(columns=['Column', 'Number of Outliers'])

for col in skewed_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    num_outliers = outliers.shape[0]
    outliers_skewed = outliers_skewed.append({'Column': col, 'Number of Outliers': num_outliers}, ignore_index=True)

# detect outliers in normal columns using Z-score
outliers_normal = pd.DataFrame(columns=['Column', 'Number of Outliers'])

for col in normal_cols:
    z_scores = zscore(df[col])
    outliers = df[np.abs(z_scores) > 3]
    num_outliers = outliers.shape[0]
    outliers_normal = outliers_normal.append({'Column': col, 'Number of Outliers': num_outliers}, ignore_index=True)

print('Skewed Columns:')
print(outliers_skewed)
print('\nNormal Columns:')
print(outliers_normal)
```
